```python
amenities.reset_index(drop=True, inplace=True)
amenities
```

```python
amenities_df = amenities.copy()
amenities_df = amenities_df.to_frame(name='string_type')
amenities_df
```

```python
# using strip and split method
```

```python
amenities_df['removed_brackets'] = amenities_df['string_type'].apply(lambda x: x.strip('[').strip(']'))
amenities_df
```

```python
amenities_df['removed_quotations'] = amenities_df['removed_brackets'].apply(lambda x: x.replace('"', ''))
amenities_df
```

```python
amenities_df['cleaned'] = amenities_df['removed_quotations'].apply(lambda x: x.split(', '))
amenities_df
```

```python
amenities_df.drop(columns= ['string_type', 'removed_brackets', 'removed_quotations'], inplace=True)
amenities_df
```
